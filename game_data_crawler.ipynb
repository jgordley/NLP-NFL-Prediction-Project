```python
def get_stats_by_week(num_weeks, week_date_map, team_abb, stat_url):
    """
    Input:
        num_weeks -> will loop from week 1 to num_weeks (inclusive) to get data
        week_date_map -> map from week num (int) to date in form 'YEAR-MONTH-DAY' of start of week
        team_abb -> map from full team name string (as shown in teamrankings.com) to abbrev in kevins dataset
        stat_url -> url of statistic to be collected from teamrankings ex. 'passing-yards-per-game'
    
    Output: 
        dataframe -> Rank	Team	2012	Last 3	Last 1	Home	Away	2011	Team_Ab Week
    """

    base_url = f'https://www.teamrankings.com/nfl/stat/{stat_url}'

    stat_list = []
    first = True

    for wk_num in range(1, num_weeks + 1):
        date = week_date_map[wk_num]
        url = base_url + f'?date={date}'

        page = requests.get(url)
        if page.status_code != 200:
            print(page.status_code)
            print(page.json())
            break

        soup = BeautifulSoup(page.content, 'html.parser')
        table = soup.find("table")

        if first:
            table_head = table.find('thead')
            head_rows = table_head.find_all('tr')
            head_cols = head_rows[0].find_all('th')
            header_vals = [h.text.strip() for h in head_cols]
            header_vals.append('Week')
            header_vals.append('Team_Abb')
            # print(header_vals)
            first = False

        table_body = table.find('tbody')
        rows = table_body.find_all('tr')
        for row in rows:
            cols = row.find_all('td')
            cols = [ele.text.strip() for ele in cols]
            cols.append(wk_num)
            cols.append(team_abb[cols[1]])
            stat_list.append(cols) # Get rid of empty values

    return header_vals, stat_list

        
        
```

```python
stats_to_collect = ['points-per-game', 'turnover-margin-per-game', 'rushing-yards-per-game']
```

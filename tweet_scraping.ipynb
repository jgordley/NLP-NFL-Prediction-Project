```python
tweet_id_to_text = collections.defaultdict(int)
```

```python
start = 0
amount = 100

tweets_received = 0

for twt_cutoff in range(100, len(tweet_ids), amount):

    # Range
    print(f'Start: {start}, End: {twt_cutoff}')

    # Form the request
    ids = map(str, tweet_ids[start:twt_cutoff])
    id_string = ','.join(ids)
    url = f'https://api.twitter.com/2/tweets?ids={id_string}&tweet.fields=created_at&expansions=author_id&user.fields=created_at'

    # Making request
    result = requests.get(url=url, headers=headers)

    # Check response
    if result.status_code != 200:
        print(result.status_code)
        print(result.json())
        break

    # Increment total number of tweets received out of 100 ids
    # some may now be private accounts or the tweets may have been deleted
    tweets_received += len(result.json().get('data'))

    for twt in result.json().get('data'):
        contents = twt['text']
        twt_index = twt['id']

        tweet_id_to_text[twt_index] = contents

    print(f'Overall percentage received: {tweets_received}/{twt_cutoff} = {(tweets_received/twt_cutoff) * 100}')

    # Cleanup for next loop
    start = twt_cutoff
```
